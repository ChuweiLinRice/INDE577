```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function for main algorithm
# input max_iter represents the number of iterations we want to perform. Alpha is the learning rate. 
# early stop represents the threshold where we stop the iterations.
def gradient_descent(x, y, max_iter = 1000, alpha = 0.0001, early_stop = 1e-6):
    # current weight and bias
    cur_weight = 0.1
    cur_bias = 0.01
    max_iter = max_iter
    alpha = alpha
    n = float(len(x))
     
    costs = []
    weights = []
    prev_cost = None
     
    for i in range(max_iter):
         
        y_pred = (cur_weight * x) + cur_bias
         
        # current cost
        cur_cost = np.sum((y - y_pred)**2) / len(y)
 
        if prev_cost and abs(prev_cost - cur_cost) <= early_stop:
            break
         
        prev_cost = cur_cost
 
        costs.append(cur_cost)
        weights.append(cur_weight)
         
        # gradients
        weight_derivative = -(2/n) * sum(x * (y - y_pred))
        bias_derivative = -(2/n) * sum(y - y_pred)
         
        # weights and bias update
        cur_weight = cur_weight - (alpha * weight_derivative)
        cur_bias = cur_bias - (alpha * bias_derivative)
                 
        # print parameters for every 1000th iteration
        print(f"Iteration {i+1}: Cost {cur_cost}, Weight {cur_weight}, \
         Bias {cur_bias}")
     
        # Visualization
    plt.figure(figsize = (8,6))
    plt.plot(weights, costs)
    plt.scatter(weights, costs, marker='o', color='red')
    plt.title("Cost vs Weights")
    plt.ylabel("Cost")
    plt.xlabel("Weight")
    plt.show()
     
    return cur_weight, cur_bias
```

# Self-generated dataset
The following is a dataset that is randomly generated to test the algorithm. It contains X and Y variable both range from 0 to 30 and have length 30. The mission is to predict the Y values after multiple iterations. I perform 100 iterations to demonstrate the algorithm. In reality the number of iterations should go higher.

```python
from numpy import random
random.seed(1)
X = random.randint(30, size=(30))
Y = random.randint(30, size=(30))
 
estimated_weight, estimated_bias = gradient_descent(X, Y, max_iter=100, alpha = 0.0001)

print(f"Estimated Weight: {estimated_weight}\nEstimated Bias: {estimated_bias}")

# predicted Y values
Y_pred = estimated_weight * X + estimated_bias
 
# Plotting the regression line
plt.figure(figsize = (8,6))
plt.scatter(X, Y, marker='o', color='blue')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red', markerfacecolor='blue',
             markersize=10,linestyle='dashed')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
 
```
